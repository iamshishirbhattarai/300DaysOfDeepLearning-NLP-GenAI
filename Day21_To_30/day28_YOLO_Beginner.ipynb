### Importing Required Libraries

```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

### Initializing Model

```python
model = YOLO('yolov8n.pt')
```

### Function For Image Detection

```python
def objectDetector(image):

  source = image
  frame = cv2.imread(source)

  if frame is None:
    print("Failed to load the image !! Please check the path !!")
  else:
    results = model(frame)

    annotated_frame = results[0].plot()

    cv2_imshow(annotated_frame)
```

### Few Examples Demonstrating Image Detection

```python
objectDetector('/content/drive/My Drive/Colab Notebooks/Images/cat_dog.jpg')
```

```python
objectDetector('/content/drive/My Drive/Colab Notebooks/Images/road.jpg')
```

```python
objectDetector('/content/drive/My Drive/Colab Notebooks/Images/woman_dog.webp')
```
